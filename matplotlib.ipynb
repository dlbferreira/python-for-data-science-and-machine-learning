# **Matplotlib**

```
import matplotlib.pyplot as plt
```

```
%matplotlib inline
```

```
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```
x
```

```
y
```

```
# FUNCTIONAL PLOT
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```
# SUBPLOT
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```
# Object Orientation
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```
# Object Orientation
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGE PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```
# Object Orientation
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
  current_ax.plot(x,y)
```

```
axes
```

```
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')

plt.tight_layout()
```

## **Figure Size and DPI**

```
fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```
fig,axes = plt.subplots(figsize=(8,2))

axes.plot(x,y)
plt.tight_layout()
```

```
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```
# Legend
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='Squared')
ax.plot(x,x**3,label='Cubed')


#ax.legend(loc=0)
ax.legend(loc=(0.1,0.5))
```

## **Plot Appearance**

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# RGB Hex Code
#ax.plot(x,y,color='#FF8C00')
ax.plot(x+1,y,color='purple')

# linewidth
ax.plot(x+2,y,color='purple',linewidth=3)

# alpha
ax.plot(x+3,y,color='purple',lw=3,alpha=0.5)

# linestyle
ax.plot(x+4,y,color='purple',lw=3,alpha=0.5,linestyle='-.')
ax.plot(x+4,y,color='purple',lw=3,alpha=0.5,ls='steps')

# marker
ax.plot(x+4,y,color='purple',lw=3,alpha=0.4,ls=':',marker='o')

# markersize
ax.plot(x+5,y,color='purple',lw=3,alpha=0.7,ls=':',marker='o', markersize=10)

# markerfacecolor
ax.plot(x+5,y,color='purple',lw=3,alpha=0.7,ls=':',marker='o', markersize=10, markerfacecolor='red')

# markeredgewidth
ax.plot(x+6,y,color='purple',lw=3,alpha=0.7,ls=':',marker='o', markersize=10, markerfacecolor='red',markeredgewidth=4)

# markeredgecolor
ax.plot(x+6,y,color='purple',lw=3,alpha=0.7,ls=':',marker='o', markersize=10, markerfacecolor='red',markeredgewidth=4,markeredgecolor='blue')
```

```
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,linestyle='--')

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
